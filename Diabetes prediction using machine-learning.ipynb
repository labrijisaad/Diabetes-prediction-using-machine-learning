> How do we **`evaluate our model ?`** <br>
> - After training the model we then apply the evaluation measures to check how the model is performing. Accordingly, we use the following evaluation parameters to check the performance of the models respectively :

> - **`Accuracy Score`** : Typically, the accuracy of a predictive model is good (above 90% accuracy)

> - **`Execution time`** : Variable that depends on the machine on which the program was executed, but which can give a small idea of the model that executes the fastest.

> - **`F1 score`** : The $F1-score$ is a weighted harmonic mean of precision and recall such that `the best score is 1.0` and `the worst is 0.0`. **F1 scores are lower than accuracy measures** as they embed precision and recall into their computation.

# 🔁  Feature Transformation :
> We observe that all the variables in the dataset except the target variable **`Outcome`** are continuous. In this section, we will perform **feature transformation** on the **continuous variables** to generate two sets of features that will be used in the subsequent **classification task**.

#### 1️⃣  Feature set 1 - **`Continuous to Categorical`**

```python
plt.figure(figsize=(18, 5))
pima['BMI'].plot()
pima['BloodPressure'].plot()
plt.legend(loc="best")
plt.show()
```

> Often when dealing with continuous variables like **BMI** or **BloodPressure**,  we may wish  to transform these **continuous variables** into **categorical variables**, which may be better predictors of **Outcome**. This transformation can be achieved as follows: 
    > - Cutting the continuous values of each feature into non-overlapping buckets.  
    > - Performing this operation on all 8 continuous variables.  
    > - Naming the resulting DataFrame as __`pima1`__. **pima1** should still contains **9 conlumns** with the same set of column names as pima.
    
> For Glucose, if we wish to adopt the **glucose** level guidelines given by **American Diabetes Association**, we can convert its values into **3** groups :
> - `Normal` $>$ $140mg$ / $dl$ of **glucose**,
> - `Prediabete` $140$ - $199mg$/$dl$ of **glucose**,
> - `Diabetes` $>$ $200mg$ / $dl$ of **glucose**

```python
pima.head(100)
```

```python
# Covertting our numeric features to categorical ones :
pima1 = pd.DataFrame()

pima1["Pregnancies"] = pd.qcut(pima["Pregnancies"], 4, labels=["low", "normal", "high", "very high"])

pima1["Glucose"] = pd.cut(pima["Glucose"], bins=[0, 140, 200, 250], right=False,
                          labels=["Normal", "Prediabetes", "Diabetes"])

pima1["BloodPressure"] = pd.cut(pima["BloodPressure"], [0, 90, 120, 140, 190], right=False, precision=0,
                                labels=["Low", "Ideal", "High", "Very high"])

pima1["SkinThickness"] = pd.cut(pima["SkinThickness"], [0, 140, 199, 250], right=False, precision=0,
                                labels=["Normal", "Prediabetes", "Diabetes"])

pima1["Insulin"] = pd.cut(pima["Insulin"], [0, 40, 60, 1000], precision=0, right=False,
                          labels=["Low", "Normal", "Diabetes"])

pima1["BMI"] = pd.cut(pima["BMI"], [0, 18.5, 24.9, 29.9, 40, 100], precision=0, right=False,
                      labels=["underweight ", "Healthy", "Overweight ", "Obese", "OverObese"])

pima1["DiabetesPedigreeFunction"] = pd.qcut(pima["DiabetesPedigreeFunction"], 4,
                                            labels=["low", "normal", "high", "very high"])

pima1["Age"] = pd.cut(pima["Age"], [21, 30, 45, 100], right=False, precision=0,
                      labels=["Young adult", "Middle-aged adult", "Old adult"])
```

# 🏗️  Model Construction :
> Our task in this section is to use the **derived feature sets** from previous section to **construct classification models** for diabetes outcome prediction (**0** or **1**).

```python
# pima1 splitting :
x1 = pima1.drop(columns='Outcome')
y1 = pima1['Outcome']
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size=0.25, random_state=0)

# pima2 splitting :
x2 = pima2.drop(columns='Outcome')
y2 = pima2['Outcome']
x_train2, x_test2, y_train2, y_test2 = train_test_split(x2, y2, test_size=0.25, random_state=0)

print(x1.shape, y1.shape)
print(x2.shape, y2.shape)
```

> Now, we are going to experiment the 2 classification models below from **`scikit-learn`** with their **default hyperparameter** settings on **`pima1`** and **`pima2`** :
> - **`KNN`**
> - **`Decision tree`**

> We will perform model training on **`x_train1`** and **`x_train2`**, and perform evaluation on **`x_test1`** and **`x_test2`** respectively, we will use **`Accuracy`** and **`weighted-F1`** as evaluation metrics.

```python
model_names = ['KNN', 'DT']
pima1_acc = []
pima2_acc = []
pima1_f1 = []
pima2_f1 = []
```

```python
# K-Nearest-Neighbors :


## Instanciating the KNN models ( K=3)
neigh1 = KNeighborsClassifier(n_neighbors=3)
neigh2 = KNeighborsClassifier(n_neighbors=3)

## Training the two models :
neigh1.fit(x_train1, y_train1)
neigh2.fit(x_train2, y_train2)

## Making prediction on test sets :
pred1 = neigh1.predict(x_test1)
pred2 = neigh2.predict(x_test2)

## Mesuring accuracy score :
acc1 = accuracy_score(y_test1, pred1)
acc2 = accuracy_score(y_test2, pred2)

## Mesuring the F1 score :
F1_score1 = f1_score(y_test1, pred1, average="weighted")
F1_score2 = f1_score(y_test2, pred2, average="weighted")

## Adding metrics to corresponding lists :
pima1_acc.append(acc1)
pima2_acc.append(acc2)
pima1_f1.append(F1_score1)
pima2_f1.append(F1_score2)
```

```python
# Decision tree :


## Instanciating the Decision Tree models :
dt1 = DecisionTreeClassifier()
dt2 = DecisionTreeClassifier()

## Training the two models :
dt1.fit(x_train1, y_train1)
dt2.fit(x_train2, y_train2)

## Making prediction on test sets :
pred1 = dt1.predict(x_test1)
pred2 = dt2.predict(x_test2)

## Mesuring accuracy score :
acc1 = accuracy_score(y_test1, pred1)
acc2 = accuracy_score(y_test2, pred2)

## Mesuring the F1 score :
F1_score1 = f1_score(y_test1, pred1, average="weighted")
F1_score2 = f1_score(y_test2, pred2, average="weighted") 

## Adding metrics to corresponding lists :
pima1_acc.append(acc1)
pima2_acc.append(acc2)
pima1_f1.append(F1_score1)
pima2_f1.append(F1_score2)
```

```python
accuracy_record = pd.DataFrame({'Model': model_names, 'pima1_acc': pima1_acc, 'pima2_acc': pima2_acc})
accuracy_record['acc_mean'] = accuracy_record.mean(axis=1).round(2)
accuracy_record.set_index('Model', inplace=True)
accuracy_record.loc['avg'] = accuracy_record.mean()

F1_record = pd.DataFrame({'Model': model_names, 'pima1_f1': pima1_f1, 'pima2_f1': pima2_f1})
F1_record['F1_mean'] = F1_record.mean(axis=1).round(2)
F1_record.set_index('Model', inplace=True)
F1_record.loc['avg'] = F1_record.mean()
print("Accuracy results :")
print(accuracy_record)
print('\n')
print("F1-score results :")
print(F1_record)
```

> - We can get deature importances for some tree-based classifiers in **`scikit-learn`** via **`clf.feature_importances_`**

> - Here is an example code for plotting **`pima2 feature importances`** for an adaboost decision tree classifier given below with a brief description how the feature importances for the given classifier can be computed.

```python
x_train = x_train2
y_train = y_train2

from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier(random_state=0)
ada.fit(x_train, y_train)
relevants = ada.feature_importances_
indices = np.argsort(relevants)[::-1]

# Plot feature importance
plt.figure(1, figsize=(9, 8))
plt.title('Feature Importance in AdaBoost :')
plt.bar(range(x_train.shape[1]), relevants[indices], color='r', align='center')
plt.xticks(range(x_train.shape[1]), x_train.columns[indices], rotation=90)
plt.xlim([-1, x_train.shape[1]])
plt.show()
```

```python
# Getting the accuracy score of the AdaBoost classifier :
Adapred = ada.predict(x_test2)
print("the accuracy of the AdaBoost Classifier is {}".format(accuracy_score(y_test2, Adapred)))
print("the f1_score of the AdaBoost Classifier is {}".format(f1_score(y_test2, Adapred)))
```

> As we can see, we have better results with the AdaBoost classifier, even better than the KNN & Decision tree models.

#### 🧐  `Observations` :
> **`BMI`** and **`DiabetesPedigreeFunction`** are more significantly **important features**, while **`Insulin`** and **`SkinThickness`** do not, which is because too many outliers in those two items.

> As for **`Adaptive boosting classifier`**, there is an attribute called `feature_importances_`. For trees, feature importance is calculated as the decrease in node impurity weighted by the probability of reaching that node. And the node probability can be calculated by the number of samples that reach the node, divided by the total number of samples.

#### ❓  `How to compute the feature importances for KNN classifiers & generate a similar bar plot showing the feature importances` ?

> The best way to **estimate feature importance** is to take the sample to predict and calculate its distance from each of its nearest neighbors for each feature (call them neighb_dist). Then do the same calculations for a few random points (call them rand_dist) instead of nearest neighbors. Then for each feature you take the ratio of neighb_dist / rand_dist, and the smaller the ratio the more important that feature is
